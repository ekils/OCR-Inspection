```
import time
from glob import glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
from langconv import Converter
import re
import model
import pandas as pd
import pytesseract
```

<font face="微软雅黑" size=5 color=	#FFD700	> ||  正式跑迴圈 || 

```
# 測試照片：
image_set = ['image01.jpg', 'image02.jpg', 'image03.jpg', 'image04.jpg', 'image05.jpg',]
paths = glob('./test/*.*')
Index = []
Disease = []
Disposal_Consideration =[]

for image_name in image_set:
    im = Image.open("./test/" + image_name)
    img = np.array(im.convert('RGB'))
    # 影像前處理： 
    ori_img = img
    lab = cv2.cvtColor(ori_img, cv2.COLOR_RGB2Lab) # 轉 Lab
    lab_plane = cv2.split(lab) 
    clahe = cv2.createCLAHE(clipLimit=1.3,tileGridSize=(13,13)) # 對比修正： 強的減低，弱的提升
    lab_plane[0] = clahe.apply(lab_plane[0])
    lab = cv2.merge(lab_plane)
    rgb = cv2.cvtColor(lab, cv2.COLOR_Lab2RGB) # 再轉回 RGB
    rgb = cv2.fastNlMeansDenoisingColored(rgb,None,10,10,7,21)  # 去燥

    # _ = plt.subplots(figsize=(7,7))
    # _ = plt.imshow(rgb)
    # _ = plt.show()
    img = rgb
    t = time.time()
    '''
    result,img,angel 分別對應：-辨識結果，檢測圖框，文字旋轉角度
    '''
    print('----> 檢測開始 <----')
    result, image, angle, text_recs = model.model(img, model='pytorch', adjust=True, detectAngle=False)
    print("It takes time:{}s".format(time.time() - t))

    _ = plt.subplots(figsize=(7,7))
    _ = plt.imshow(image)
    _ = plt.show()

    print('原圖大小：',img.shape)
    print("---------------------------------------")

    dic = result
    rule = re.compile(r"[^a-zA-Z0-9\u4e00-\u9fa5]")
    # 非必要文字區塊篩選I: 取中間位置，因為上面是基本資料， 下面是簽名相關
    if img.shape[0] > 900 and img.shape[0] <1200:
        y_min = 100
    else:
        y_min = 310

    all_location = []
    y_ = 0
    s_ = ''
    if image_name == 'image03.jpg':
        margin_pix = 60
    else:
        margin_pix = 40
    # 非必要文字區塊篩選II: 如果是兩行文字，彼此檢測匡的y距離會很相近，假如超過40pix 則認為是不同區塊
    for key in dic:
        y_string = dic[key][1]
        y_location = dic[key][0].tolist()[1]
        y_string = Converter('zh-hant').convert(y_string)
        y_string = rule.sub('',y_string)
        if len(y_string) > 5 and y_location > y_min and y_location < 700:
            if (abs(y_ - y_location) <= margin_pix and abs(y_ - y_location) > 10 ): # 介於 10~ margin_pix之間的距離才認定是同一區塊
                all_location.append((s_,y_))
            else:
                all_location.append((s_,y_))
                all_location.append('x')
            y_ = y_location
            s_ = y_string
    _ = all_location.pop(0)
    # all_location

    # 將上面篩選出來的結果做group：
    count = 0
    prob_index = 0
    probably_discription = []
    for i in all_location:
        if i =='x':
            count += 1
        else:
            probably_discription.append((prob_index,i))
            count = 0
        if count > 0:
            prob_index += 1
    temp_index = list(set([i[0] for i in probably_discription]))

    # 看有幾個group 就創造幾個空list[]：
    obj = [[] for i in range(len(temp_index))]
    for i in range(len(temp_index)):
        for index in probably_discription:
            if index[0] == temp_index[i]:
                obj[i].append(index[1])

    # 診斷內容寫入：
    if image_name == 'image01.jpg':
        temp_disease = obj[1]
    else:
        temp_disease = obj[0]
    disease = ''
    for i in temp_disease: 
        d = (i[0]) + '\n'
        disease = disease + d
    print(disease)

    # 醫生叮囑寫入：
    if image_name == 'image01.jpg':
        temp_disposal = obj[2]
    else:
        temp_disposal = obj[1]
    disposal = ''
    for i in temp_disposal: 
        d = (i[0]) + '\n'
        disposal = disposal + d
    print(disposal)
    Index.append(image_name)

    Disease.append(disease)
    Disposal_Consideration.append(disposal)

# 寫入 pandas：
a = pd.Series(Index)
b = pd.Series(Disease)
c = pd.Series(Disposal_Consideration)

d = {'Index':a,'Disease': b, 'Disposal_Consideration': c}
df = pd.DataFrame(data=d)
df
```

```
# 寫入 CSV:
with open ('./csv/nlptest.csv','a') as f:
    df.to_csv(f,index=False,header=False)
```
